```python
# test data
db = ''
terms= ['breast cancer']
year='2008'
journal='Science'
recursiveSearch=0
```

```python
# generate request from terms, year, etc.
def generateRequest():
  query = f'esearch.fcgi?'
  if db:
    query = query + f'db={db}'
  termsConcat = '+'.join(terms).strip(' ').replace(' ', '+')
  query = query + f'&term={termsConcat}'
  if journal:
    query = query + f'+AND+{journal}[journal]'
  if year:
    query = query + f'+AND+{year}[pdat]'
  return query
```

```python
try:
  query = generateRequest()
  print(base_search_url + query)
  result = requests.get(base_search_url + query)
except:
  print('error')
else:
  soup = BeautifulSoup(result.text, 'lxml')
```

```python
# get Ids of articles
idArticles = [p.text for p in soup.find_all('id')]
idArticles
```

```python
base_pubmed_url = 'https://pubmed.ncbi.nlm.nih.gov/'

for id in idArticles: 
  try:
    result = requests.get(base_pubmed_url + id)
  except:
    print(f'error: {id}')
  else:
    soup = BeautifulSoup(result.text, 'lxml')
```

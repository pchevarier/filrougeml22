```python
import requests
from bs4 import BeautifulSoup as bs
import json
import re
```

```python
# GetHTML_PMC
PMCID = "PMC6702738"
URL = f"https://www.ncbi.nlm.nih.gov/pmc/articles/{PMCID}/"

r = requests.get(URL)
if r.status_code != 200:
    print("PMC blocked")
else:
    html_PMC = bs(r.text, 'html.parser')

```

```python
# Get_PMID_of_references

PMIDs_cited = []
references = html_PMC.find('div', {'id': "reference-list"}).find_all(href=re.compile("pubmed"))
for reference in references:
    PMID_ = reference['href'].replace('/pubmed/', '')
    PMIDs_cited.append(PMID_)
PMIDs_cited_dict = {'PMIDs_cited': PMIDs_cited}
print(PMIDs_cited_dict)
    

```

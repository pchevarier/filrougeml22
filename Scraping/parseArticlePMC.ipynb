```python
import requests
from bs4 import BeautifulSoup as bs
import json
```

```python
# GetHTML_PMC
URL = "https://www.ncbi.nlm.nih.gov/pmc/articles/PMC359599/"

r = requests.get(URL)
print(f"Status code: {r.status_code}")
html_paper = bs(r.text, 'html.parser')
```

```python
# GetDatePMC
journal_date_issue = html_paper.find('div', {'class': "part1"}).text
date = journal_date.split('.')[1].split(';')[0].strip()
date
# a retravailler
```

```python
# GetTitlePMC
title = html_paper.find('h1', {'class': "content-title"}).text
print(title)
```

```python
# GetDOIPMC
DOI = html_paper.find('span', {'class': "doi"}).a['href']
only = DOI.replace("//dx.doi.org/", '')
only
```

```python
# GetAbstractPMC
abstract = html_paper.find('p', {'class': "p p-first-last"}) 
print(abstract.text)
```

```python
# GetRefsPMC
references = html_paper.find_all('span', {'class': "nowrap"})
for i in references:
    print(i.a['href'])
```

```python
# Get_FulltextPMC
all = html_paper.find_all('div', {'class': "tsec sec"}) 
for i in all:
    print(i.text)
```

```python
# GetAuthorsPMC
authors = html_paper.find_all('a', {'class': 'affpopup'})
for author in authors:
    print(author.text)
```

```python
# Affiliations
all = html_paper.find('div', {'class': 'contrib-group fm-author'})
authors = all.find_all('a')
for author in authors:
    print(author.text)
numbers = all.find_all('sup')



```
